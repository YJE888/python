#### **긴 문장 자르기 - textwrap.wrap()**

```python
# 리스트 형식으로 40자씩 묶어주게 됨
wrapped_content = textwrap.wrap(content, width=40)
```

```python
print(wrapped_content)
```

```python
# wrapped_content에서 요소들을 줄바꿈해서 표현
print('\n'.join(wrapped_content))
```

#### **textwrap.fill()**
- 상단에서 리스트 형식으로 묶고 줄바꿈으로 표현하는 두 단계를 하나의 단계로 사용하게 해주는 함수

```python
filled_content = textwrap.fill(content, width=40)
```

```python
print(filled_content)
```

## **문자열 다루기 - re**
> - 정규표현식(regular expressions)은 복잡한 문자열을 처리할 때 사용하는 기법으로, 파이썬뿐 아니라 C, 자바, 심지어 문서 작성 프로그램 등 문자열을 처리해야 하는 다양한 곳에서 활용중
> - 특정 문자열 `추출`, `변환` 등에 사용
> - [공식문서](https://docs.python.org/3/library/re.html)

#### **re.findall()**
> - 단어 추출

```python
import re
```

```python
# content의 단어 추출
words = re.findall(r'\w+', content)
```

```python
print(words)
```

#### **전화번호 추출**

```python
# 연락처 목록
contact = '''원숭이 21 010-1234-5566 Monkey@google.com 
             고양이 20 010-5678-9900 cat@naver.com
             강아지 34 010-2039-4958 dog@school.ac.com
             망아지 40 010 7400 5528 Horse@school.ac.com
             송아지 23 010-7777-2232 cow@google.com'''
```

```python
# 전화번호 추출
# 0으로 시작하고, 숫자(d)로 시작하면서 한자리 또는 두자리이고, - , 숫자(d)로 시작하면서 세자리 또는 네자리이고, -, 숫자(d)로 시작하면서 세자리 또는 네자리인 수
regex = r'0\d{1,2}[ -]?\d{3,4}[ -]?\d{3,4}'
 
phone = re.findall(regex, contact)
print("\n".join(phone))
```

#### **문자열 변환**

```python
# 변환(마스킹)
pat = re.compile(r'0\d{1,2}[ -]?\d{3,4}[ -]?\d{3,4}')
print(pat.sub("010-****-****", contact))
```
